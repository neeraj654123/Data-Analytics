# Plotly Express

```python
import plotly.express as px
```

```python
px.data.stocks()
```

```python
df=px.data.stocks()
df.info()
```

```python
px.line(df,y='GOOG',x='date')
```

```python
px.line(df,y=['GOOG',"FB"],x='date')
```

```python
px.line(df,y=['GOOG',"FB"],x='date',range_x=["2019","2019-09"])
```

### Labels and Titles

```python
px.line(df,y=['GOOG',"FB"],x='date',labels={'value':'close','date':'time'},title='Stocks')
```

### Height and Width

```python
px.line(df,y=['GOOG',"FB"],x='date',height=500,width=1000)
```
