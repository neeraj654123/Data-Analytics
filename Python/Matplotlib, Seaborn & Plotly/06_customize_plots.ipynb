```python
heart_disease =pd.read_csv("C:/Users/asus/OneDrive/ドキュメント/Tutedude/Data Analytics/Python/Matplotlib, Seaborn & Plotly/heart-disease.csv")
over_50 =heart_disease[heart_disease["age"]>50]
over_50
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
# Creating Figure
fig,ax=plt.subplots(figsize=(10,7))
# Creating Plot
scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter")
# Customize Plot
ax.set(title="Heart Disease and Cholestrol Level",xlabel="Age",ylabel="Cholestrol")
# Add Legend
ax.legend(*scatter.legend_elements(),title="target")
# Add a horizontal line
ax.axhline(over_50.chol.mean(),ls="--")
```

```python
# Subplot of age, chol, thalach
fig,(ax0,ax1)=plt.subplots(figsize=(10,10),nrows=2,ncols=1)
# Customize and Adding data to ax0
scatter =ax0.scatter(x=over_50.age,y=over_50.chol,c=over_50.target,cmap="winter")
ax0.set(title="Heart Disease and Cholestrol Level",xlabel="Age",ylabel="Cholestrol")
ax0.set_xlim(50,80)
ax0.legend(*scatter.legend_elements(),title="target")
ax0.axhline(over_50.chol.mean(),ls="--")
# Customize and Adding data to ax1
scatter =ax1.scatter(x=over_50.age,y=over_50.thalach,c=over_50.target,cmap="winter")
ax1.set(title="Heart Disease and Thalach (Max Heart Rate)",xlabel="Age",ylabel="Thalach")
ax1.set_xlim(50,80)
ax1.set_ylim(60,200)
ax1.legend(*scatter.legend_elements(),title="target")
ax1.axhline(over_50.thalach.mean(),ls="--")
# Adding a Super Title
fig.suptitle("Heart Disease Analysis",fontsize=15,fontweight="bold")
```
