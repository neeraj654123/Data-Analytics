# Histograms and Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x =np.linspace(0,10,100)
x[:20]
```

### Histograms Plot

```python
data =np.random.randn(1000)
fig,ax=plt.subplots(figsize=(5,5))
ax.hist(data)
```

# Subplots
##### Method 1

```python
butter_prices ={"Amul":58,
                "Verka":65,
                "Urmul":50,
                "Delicious":25,}
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x,x**2)
ax2.set(title = "Butter Prices of different Brands",
      xlabel="Brands",
      ylabel="Butter Prices in Rs(₹)")
ax2.bar(butter_prices.keys(),butter_prices.values())
ax3.barh(butter_prices.keys(),butter_prices.values())
ax4.hist(data)
```

##### Method 2

```python
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x**2)
ax[0,1].set(title = "Butter Prices of different Brands",
      xlabel="Brands",
      ylabel="Butter Prices in Rs(₹)")
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(butter_prices.keys(),butter_prices.values())
ax[1,1].hist(data)
```
