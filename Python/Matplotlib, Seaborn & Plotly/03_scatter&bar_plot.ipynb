# Scatter and Bar plot


```python
import matplotlib.pyplot as plt
import numpy as np
```

### Making Figures using Numpy Arrays

```python
x =np.linspace(0,10,100)
x[:20]
```

```python
# Simple Line plot
fig,ax=plt.subplots(figsize=(5,5))
ax.plot(x,x**2)
```

#### Scatter Plot

```python
fig,ax=plt.subplots(figsize=(5,5))
ax.scatter(x,x**2)
```

#### Bar Plot

```python
butter_prices ={"Amul":58,
                "Verka":65,
                "Urmul":50,
                "Delicious":25,}
fig,ax=plt.subplots(figsize=(5,5))
ax.set(title = "Butter Prices of different Brands",
      xlabel="Brands",
      ylabel="Butter Prices in Rs(₹)")
ax.bar(butter_prices.keys(),butter_prices.values())
```

```python
# Horizontal Bar Plot
fig,ax=plt.subplots(figsize=(5,5))
ax.set(title = "Butter Prices of different Brands",
      xlabel="Brands",
      ylabel="Butter Prices in Rs(₹)")
ax.barh(butter_prices.keys(),butter_prices.values())
```
