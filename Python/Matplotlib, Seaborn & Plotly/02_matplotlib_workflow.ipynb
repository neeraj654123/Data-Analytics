```python
# 1. Import matplotlib
import matplotlib.pyplot as plt
```

```python
# 2. Prepare data
x=[1,2,4,8]
y=[11,22,33,44]
# 3. Setup plot
fig,ax=plt.subplots(figsize =(5,5))
# 4. Plot data
ax.plot(x,[50,180,250,500])
# 5. Customize the figure
ax.set(title = "Simple plot",
      xlabel="X-Axis",
      ylabel="Y-Axis")
# 6. Save the plot
fig.savefig("Plot.png")
```
