```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Stateful Method:
##### Figure and Axes are created and managed automatically by matplotlib.It is Best for simple and quick plotting.

```python
# Ploting graph without any data
plt.plot()
```

```python
# Ploting graph with data
x=[1,2,4,8]
y=[11,22,44,55]
plt.plot(x,y)
```

# Explicit Method (Object-Oriented-Interface) :
##### This method requires you to manually create and manage Figure and Axes objects, giving greater control and flexibility for complex visualizations

```python
# 1st Method of explicit 
fig =plt.figure()
ax =fig.add_subplot()
```

```python
# 2nd Method
fig = plt.figure()
ax =fig.add_axes([1,1,1,1])
ax.plot(x,y)
```

```python
# 3rd Method (recommended)
fig,ax=plt.subplots()
ax.plot(x,[50,180,250,500])
```
