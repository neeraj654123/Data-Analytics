```python
car_data =pd.read_csv("C:/Users/asus/OneDrive/ドキュメント/Tutedude/Data Analytics/Python/Pandas/car-sales.csv")
car_data
```

```python
car_data["Price"] = car_data.Price.str.replace('[\$\,\.]',"",regex=True)
car_data["Price"] = car_data.Price.str[:-2] # It removes the last 2 characters from every value
car_data["Sales Date"]=pd.date_range("1/1/2026",periods=len(car_data))
car_data["Total Sales"] =car_data["Price"].astype(int).cumsum()
car_data
```

```python
heart_disease =pd.read_csv("C:/Users/asus/OneDrive/ドキュメント/Tutedude/Data Analytics/Python/Matplotlib, Seaborn & Plotly/heart-disease.csv")
heart_disease.head(10)
```

```python
# Histogram of age
heart_disease.age.plot.hist(bins=20)
```

```python
heart_disease.plot.hist(subplots=True,figsize=(10,50))
```

```python
over_50 =heart_disease[heart_disease["age"]>50]
over_50
```

```python
# Implicit Method
over_50.plot(kind="scatter",x="age",y="chol",c="target")
```

```python
# Explicit Method mixed with Stateful method
fig,ax=plt.subplots()
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax,figsize=(10,7))
ax.set_xlim(40,100) # It set X-axis limit
```

```python
## Pure Explicit Method
# Creating Figure
fig,ax=plt.subplots(figsize=(10,7))
# Creating Plot
scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
# Customize Plot
ax.set(title="Heart Disease and Cholestrol Level",xlabel="Age",ylabel="Cholestrol")
# Add Legend
ax.legend(*scatter.legend_elements(),title="target")
# Add a horizontal line
ax.axhline(over_50.chol.mean(),ls="--")
```

```python
# Subplot of age, chol, thalach
fig,(ax0,ax1)=plt.subplots(figsize=(10,10),nrows=2,ncols=1)
# Customize and Adding data to ax0
scatter =ax0.scatter(x=over_50.age,y=over_50.chol,c=over_50.target)
ax0.set(title="Heart Disease and Cholestrol Level",xlabel="Age",ylabel="Cholestrol")
ax0.legend(*scatter.legend_elements(),title="target")
ax0.axhline(over_50.chol.mean(),ls="--")
# Customize and Adding data to ax1
scatter =ax1.scatter(x=over_50.age,y=over_50.thalach,c=over_50.target)
ax1.set(title="Heart Disease and Thalach (Max Heart Rate)",xlabel="Age",ylabel="Thalach")
ax1.legend(*scatter.legend_elements(),title="target")
ax1.axhline(over_50.thalach.mean(),ls="--")
# Adding a Super Title
fig.suptitle("Heart Disease Analysis",fontsize=15,fontweight="bold")
```
