# Seaborn Univariate Analysis

```python
import pandas as pd
import seaborn as sb
import numpy as np
```

```python
data =sb.load_dataset('exercise')
data
```

```python
sb.histplot(data=data,x='pulse')
```

```python
sb.boxplot(data=data,x='pulse')
```

```python
sb.boxenplot(data=data,x='pulse')
```

```python
sb.violinplot(data=data,x='pulse')
```

```python
sb.rugplot(data=data,x='pulse')
```

```python
sb.kdeplot(data=data,x='pulse',fill=True,color='blue')
```

```python
sb.stripplot(data=data,x='pulse')
```

```python
sb.swarmplot(data=data,x='pulse')
```

```python
sb.ecdfplot(data=data,x='pulse',hue='diet')
```

### Combining Univariate Charts

```python
sb.kdeplot(data=data,x='pulse',hue='diet')
sb.rugplot(data=data,x='pulse',hue='diet')
```

# Seaborn Bivariate Analysis

```python
data =sb.load_dataset('flights')
data
```

```python
sb.lineplot(data=data,x="year",y="passengers")
```

```python
sb.pointplot(data=data,x="year",y="passengers")
```

```python
sb.barplot(data=data,x="year",y="passengers",color='black')
```

```python
data =sb.load_dataset('diamonds')
data
```

```python
data =data.sample(3000)
```

```python
sb.scatterplot(data=data,x="carat",y='price',marker='+')
```

```python
sb.kdeplot(data=data,x="carat",y='price',fill=True,cmap="mako_r")
```

```python
sb.histplot(data=data,x="carat",y='price')
```

```python

sb.histplot(data=data,x="carat",y='price',color='red')
sb.kdeplot(data=data,x="carat",y='price',fill=True,cmap="coolwarm")
```

```python

sb.jointplot(data=data,x="carat",y='price',color='red',kind="hex")
```

```python
sb.lmplot(data=data,x="carat",y='price',line_kws={'color':'red'})
```

# Seaborn Plotting Relationships

```python
sb.relplot(data=data,x="carat",y='price',col='cut')
```
