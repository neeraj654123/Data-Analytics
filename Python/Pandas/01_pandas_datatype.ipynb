# Pandas Datatype

```python
import pandas as pd
```

##### Series : 1 Dimensional

```python
cars =pd.Series(["Nissan", "Volvo", "Toyota", "Honda"]);
colours =pd.Series(["yellow", "green", "black", "orange"])
cars, colours
```

```python
# Indexing
cars[2] , cars[0]
```

##### Data Frames : 2 Dimensional

```python
car_data = pd.DataFrame({"Brand":cars, "Colour":colours})
car_data
```

##### Importing a CSV file and turning it into a data frame

```python
cars_data = pd.read_csv("car-sales.csv")
cars_data
```

##### Exporting data to a new CSV file

```python
car_data.to_csv("car_data_exported.csv")
```
