# Data Manipulation

```python
import pandas as pd
import numpy as np
```

```python
cars_data = pd.read_csv("car-sales.csv")
cars_data
```

##### Convert whole data in Make column into lowercase 

```python
cars_data["Make"]=cars_data["Make"].str.lower()
cars_data
```

```python
car_missing = pd.read_csv("car-sales-missing-data.csv")
car_missing
```

##### Filling missing value in car_missing

```python
car_missing["Odometer"]=car_missing["Odometer"].fillna(car_missing["Odometer"].mean())
car_missing
```

##### This will drop all the rows containing any missing value

```python
car_missing.dropna(inplace=True)
car_missing
```

##### Saving data whose all columns are filled as car_data_dropped.csv

```python
car_missing = pd.read_csv("car-sales-missing-data.csv")
car_missing_dropped=car_missing.dropna()
car_missing_dropped.to_csv("car_data_dropped.csv")
car_missing_dropped
```

##### Adding New Column in cars_data

```python
# Using Series
seat_column = pd.Series([5,7,5,7,5,7])
cars_data["No. of Seats"]=seat_column
cars_data
```

```python
cars_data["No. of Seats"].fillna(5,inplace=True)
cars_data
```

```python
# Using Python List
Mileage=[16.9,15,19,13,8,17,15,16.6,20,15.9]
cars_data["Mileage"]=Mileage
cars_data
```

```python
cars_data["Fuel Used"]=(cars_data["Odometer (KM)"]/cars_data["Mileage"])
cars_data
```

```python
# Create column with single value
cars_data["Wheels"] = 4
cars_data
```

##### Droping column

```python
cars_data.drop("Wheels",axis=1,inplace=True)
cars_data
```

### Apply Function

```python
cars_data["Odometer (KM)"]=cars_data["Odometer (KM)"].apply(lambda x : x / 1.6)
cars_data
```
