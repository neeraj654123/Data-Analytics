# Sorting Arrays

```python
np.random.seed(3)
random_array = np.random.randint(10,size=(5,5))
```

```python
random_array
```

```python
np.sort(random_array)
```

```python
# Returns the indices that would sort each column of the array in ascending order
np.argsort(random_array,axis=0)
```

```python
# Finds the index of the maximum value row-wise
np.argmax(random_array,axis=1)
```

##### Note : axis = 0  → Columns ↓
#####        axis = 1  → Rows →
